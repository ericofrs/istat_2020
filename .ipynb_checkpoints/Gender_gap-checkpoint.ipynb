```python
# Set version, format and language
VERSION = 'v2.1'
FORMAT = 'json'
LANGUAGE = 'en'
# call client
client = EurostatAPIClient(VERSION, FORMAT, LANGUAGE)
```

```python
# Set colors
c_tot = '#f4c95d'
c_mal = '#f36744'
c_fem = '#134f67'

# style
plt.style.use('fivethirtyeight')
```

```python
# Set parameters #AGE == 25~64
params = {
#    'sinceTimePeriod':'2013',
    'precision':'2'
}


# ED0-2 = Less than primary, primary and lower secondary education (levels 0-2)
# ED3-8 = Upper secondary, post-secondary non-tertiary and tertiary education (levels 3-8)
# ED3_4 = Upper secondary and post-secondary non-tertiary education (levels 3 and 4)
# ED5-8 = Tertiary education (levels 5-8)

dic = {'time' : 'year'}

# GET the data
dataset = client.get_dataset('earn_gr_gpgr2', params=params)
#print(dataset.label)


# Transform into a dataframe
df = dataset.to_dataframe()
df.rename(columns=dic, inplace=True)
df['year'] = df['year'].astype(int)
print(df)
```

```python
df = df[df['year']==2017]
keep = ['values', 'geo', 'nace_r2', 'year']
df = df[keep]
df.pivot_table(index=['geo','nace_r2'], columns=['year'], values='values', aggfunc='last')

pivoted = df.pivot_table(index=['geo','nace_r2'], columns=['year'], values='values', aggfunc='last')

gap17 = pivoted.reset_index()
```
